## Calgary Weather

```python
# Import 3rd party libraries
import os
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
#Viwing weather data
csv_file_path = os.path.join('Calgary_intl_weather','en_climate_daily_AB_3031094_2019_P1D.csv')
calgary_weather = pd.read_csv(csv_file_path)
calgary_weather.info()
```

```python
#Removing categories with no data for simplicity
calgary_weather_reduced = calgary_weather.drop(columns=['Data Quality', 'Total Rain (mm)','Total Rain Flag','Total Snow (cm)', 'Total Snow Flag', 'Snow on Grnd Flag'])
```

```python
#Check to make sure data is complete and makes sense
calgary_weather_reduced.plot(x='Date/Time', y='Total Precip (mm)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Convert combined dataframe into a csv 
calgary_weather_reduced.to_csv('calgary_weather_simplified.csv', index='False')
```

## Montreal Weather

```python
#Create a list of file names
montreal_weather_files = [os.path.join('Montreal_weather', file) for file in os.listdir('Montreal_weather') if 'en_climate_daily_QC_7024745' in file]

# Loop through the selected weather files, create individual dataframes for each sheet, and add dfs to a list 
dfs=[]
for file in montreal_weather_files:
    temp_df = pd.read_csv(file)
    dfs.append(temp_df)
    
# Concatenate all individual DataFrames from the list into a single DataFrame
montreal_weather = pd.concat(dfs, ignore_index=True)
```

```python
montreal_weather.info()
```

```python
#Remove columns with no info and sort Date/Time to avoid graphs with data in the wrong order
montreal_weather_reduced = montreal_weather.drop(columns=['Data Quality', 'Total Rain (mm)','Total Rain Flag','Total Snow (cm)', 'Total Snow Flag', 'Snow on Grnd Flag']).sort_values(by='Date/Time')
```

```python
#Verify that data is complete and makes sense
montreal_weather_reduced.plot(x='Date/Time', y='Total Precip (mm)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Convert combined dataframe into a csv 
montreal_weather_reduced.to_csv('montreal_weather_simplified.csv', index='False')
```
